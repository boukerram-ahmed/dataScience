```python
month_order = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
day_order = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
```

```python
def get_month(x):
    try:
        return month_order[int(str(x).split('-')[1]) - 1]
    except:
        return np.nan
```

```python
def get_day(x):
    try:
        year, month, day = (int(i) for i in x.split('-'))    
        
        answer = datetime.date(year, month, day).weekday()
        return day_order[answer]
    except:
        return np.nan
```

```python
def get_director(x):
    for i in x:
        if i['job'] == 'Director':
            return i['name']
    return np.nan
```

```python
def get_actor(x):
    act=[]
    for acteur in x:
        act.append(acteur['name'])
    return act
def get_actorArray(x):
    ligne_acteur=np.zeros(len(actors))
    for acteur in x:
        ligne_acteur[actors[acteur]]=1
    return ligne_acteur
```
